# Bybit perpetual futures leverage

List of leverage for perpetual futures on exchange Bybit

![purple-divider](resources/purple_divider.png)

## Hands on! 

```python
import pandas as pd
import requests
```

![green-divider](resources/green_divider.png)

## Loading data

```python
INSTRUMENTS_INFO_URL = "https://api-testnet.bybit.com/derivatives/v3/public/instruments-info?category=linear"

instruments_info_raw = requests.get(INSTRUMENTS_INFO_URL).json()
```

![green-divider](resources/green_divider.png)


## The data at a glance

```python
str(instruments_info_raw)[:500]
```

![green-divider](resources/green_divider.png)


## Transform data

```python
instruments_info = pd.DataFrame([{"ticker" : x["symbol"], "maxLeverage" : float(x["leverageFilter"]["maxLeverage"]) } for x in instruments_info_raw["result"]["list"] if "USDT" in x["symbol"]])

instruments_info.sort_values("maxLeverage", ascending=False).head(50)

```

![green-divider](resources/green_divider.png)


## Statistic

### Instruments with leverage 100x and higher

```python
instruments_info[instruments_info["maxLeverage"] >= 100].sort_values("maxLeverage", ascending=False)
```

### Instruments with leverage 50x and higher

```python
selected_instruments = instruments_info[instruments_info["maxLeverage"] >= 50].sort_values("maxLeverage", ascending=False)

selected_instruments.head(20)
```

```python
selected_instruments.shape[0]
```

### Instruments with leverage 25x and higher

```python
selected_instruments = instruments_info[instruments_info["maxLeverage"] >= 25].sort_values("maxLeverage", ascending=False)

selected_instruments.head(20)
```
